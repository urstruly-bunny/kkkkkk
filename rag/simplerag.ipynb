```python
from langchain_community.document_loaders import WebBaseLoader
import bs4
#load,chunk ,index the content of html page
loader=WebBaseLoader(web_paths=("https://www.filmibeat.com/celebs/mahesh-babu/biography.html",), bs_kwargs=dict(parse_only=bs4.SoupStrainer(
    id=("wrapper")
)))
web_doc=loader.load()
web_doc
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("assign.pdf")
doc=loader.load()
print(doc)
```

```python
#transform into chunks

from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
document=text_splitter.split_documents(doc)
document[:5]
```

```python
#vector embedding and vector store

from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
db= FAISS.from_documents(document[:5],OllamaEmbeddings(model="gemma:2b"))
```

```python
query="Agility & Adaptability"
result=db.similarity_search(query)
result
```
